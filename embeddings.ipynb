```python
#Listagem de documentos que serão buscados
DOCUMENT1 = {
    "Título": "Infraestrutura de TI",
    "Conteúdo": "Os bancos utilizam servidores robustos e sistemas de armazenamento para gerenciar grandes volumes de dados. Esses servidores hospedam os sistemas de banco de dados que armazenam informações importantes sobre contas, transações, clientes, etc. "}

DOCUMENT2 = {
    "Título": "Sistemas de Banco de Dados",
    "Conteúdo": "O coração do sistema operacional de um banco é o banco de dados, que armazena todas as informações financeiras e de clientes. Esses sistemas são projetados para serem altamente seguros e eficientes, permitindo acesso rápido e seguro aos dados."}

DOCUMENT3 = {
    "Título": "Aplicações Bancárias",
    "Conteúdo": "Incluem softwares que gerenciam operações diárias, como processamento de transações, gerenciamento de contas, concessão de empréstimos, etc. Essas aplicações são acessadas tanto pelos funcionários do banco, quanto pelos clientes através das plataformas online e pelos aplicativos móveis."}

DOCUMENT4 = {
    "Título": "Segurança",
    "Conteúdo": "A segurança é uma prioridade máxima. Os bancos implementam várias camadas de segurança, incluindo criptografia de dados, autenticação multifator e monitoramento contínuo para proteger contra fraudes e ataques cibernéticos."}

DOCUMENT5 = {
    "Título": "Integração com o Banco Central",
    "Conteúdo": "Os bancos comerciais operam sob a supervisão do Banco Central, que regula e monitora suas atividades para garantir a estabilidade do sistema financeiro. Isso inclui a gestão de reservas obrigatórias, controle de crédito."}

documents = [DOCUMENT1, DOCUMENT2, DOCUMENT3, DOCUMENT4, DOCUMENT5]
```

```python
df = pd.DataFrame(documents)
df.columns = ["Título", "Conteúdo"]
df
```

```python
model = "models/embedding-001"
```

```python
def embed_fn(title, text):
  return genai.embed_content(model=model,
                             content=text,
                             title=title,
                             task_type="RETRIEVAL_DOCUMENT")["embedding"]
```

a função apply é para aplicar a estrutura criada da função acima, que está gerando os embeddings, para cada linha que tenha um DataFrame (é um método dos DataFrames pandas)

```python
df["embeddings"] = df.apply(lambda row: embed_fn(row["Título"], row["Conteúdo"]), axis=1)
df
```

Calcular a distância das palavras e as similaridades


```python
def gerar_e_buscar_consulta(consulta, base, model):
  consulta_embedding = genai.embed_content(model=model,
                                            content=consulta,
                                            task_type="RETRIEVAL_QUERY")["embedding"]
  produtos_escalares = np.dot(np.stack(df["embeddings"]), consulta_embedding)

  indice = np.argmax(produtos_escalares)
  return df.iloc[indice]["Conteúdo"]
```

```python
consulta = "O banco de dados do banco pode me proteger contra ataques cibernéticos?"

trecho = gerar_e_buscar_consulta(consulta, df, model)
print(trecho)
```
